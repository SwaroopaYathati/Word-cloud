```python
input_words = ("innovation, technology, data, science, future, learning,research, development, analytics, growth, solutions, creativity, discovery,progress, insight, collaboration, trends, advancement, success, strategy")
```

```python
input_words
```

# import the worldcloud class from the worldcloud library
# import the pyplot module from the matplotlib library

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(width=480, height=400, margin=2).generate(input_words)
```

```python
plt.imshow(wordcloud, interpolation='bicubic')
plt.axis('off')
plt.margins(x=0, y=0)
plt.show()
```

```python
input_words2 = ("imagination, art, design, inspiration, colors, creativity,expression, passion, style, vision, innovation, aesthetics, ideas,originality, beauty, dynamic, conceptual, technique, craft, style")
```

```python
wordcloud2 = WordCloud(width=700, height=500, margin=50).generate(input_words2)
```

```python
plt.imshow(wordcloud2, interpolation='bicubic')
# interpolation='bicubic' -- This method uses the 16 nearest pixels (a 4x4grid) to estimate the color of each pixel.
plt.axis('on') # Shows axis
plt.margins(x=20, y=10) # horizontal and vertical margin (padding)
plt.show()
```
